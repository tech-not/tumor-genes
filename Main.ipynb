```python
def search_for_tsvs(dir_to_search):
    for _dir, subdirs, files in os.walk(dir_to_search):
        for file in files:
            if file.split('.')[-1] == 'tsv':
                yield _dir + '/' + file
```

```python
_dir = 'pancreas_transcriptomes'
format = '# gene-model: GENCODE v36'
```

```python
def get_genes(filename):
    with open(filename, 'r') as file:
        lines = list(file.readlines())
        if lines[0] != '# gene-model: GENCODE v36\n':
            print('Wrong format, cannot process', filename)
            return
        lines = lines[6:]
        
        genes = []
        for line in lines:
            cols = line.split()
            genes.append((cols[1], cols[6]))
    return genes
```

```python
for filename in search_for_tsvs(_dir):
    genes = get_genes(filename)
    print(max(genes, key=lambda a: a[1]))
```
